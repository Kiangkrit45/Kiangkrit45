### Import Library

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

### Read csv

อ่านข้อมูลทั้ง 3 ไฟล์มาเก็บไว้ในตัวแปร

```python
df_gender = pd.read_csv('data set/titanic/gender_submission.csv', sep = ',', encoding ='utf-8')

df_test = pd.read_csv('data set/titanic/test.csv', sep = ',', encoding ='utf-8')

df_train = pd.read_csv('data set/titanic/train.csv', sep = ',', encoding ='utf-8')
```

### 1. Merge data

#### รวมข้อมูลทั้งหมดให้อยู่ในไฟล์เดียวกัน

```python
df_test_merge = pd.merge(df_gender,df_test, on='PassengerId')

df_raw_all = pd.concat([df_train, df_test_merge] ,axis = 0)
```

```python
df_raw_all
```

### 2. Histogram
#### 2.1 สร้างกราฟ histogram ตามข้อมูลอายุ(Age) โดยรวมข้อมูลทั้งผู้รอดชีวิตและไม่รอดชีวิต

### Missing data

```python
df_raw_all.isnull().sum()
```

```python
df_miss_all = df_raw_all.dropna(subset=['Age'], axis=0)
```

```python
df_miss_all
```

```python
sns.distplot(df_miss_all.Age)
```

#### 2.2 สร้างกราฟ histogram ตามข้อมูลอายุ(Age) เฉพาะผู้รอดชีวิตเท่านั้น

```python
sns.distplot(df_miss_all[df_miss_all['Survived'] == 1].Age)
```

### 3. Bar Chart
#### 3.1  สร้างกราฟ barplot นับจำนวนคนทั้งหมด แบ่งตามเพศ(Sex)

```python
sns.countplot(x='Sex', data=df_raw_all)
```

#### 3.2 สร้างกราฟ barplot นับจำนวนคน แบ่งตามผู้รอดชีวิตและไม่รอดชีวิตและแบ่งตามเพศ(Sex)

```python
sns.countplot(x='Sex', hue = 'Survived', data=df_raw_all)
```

#### 3.3 สร้างกราฟ barplot นับจำนวนคน แบ่งตามผู้รอดชีวิตและไม่รอดชีวิตและแบ่งตามชั้นโดยสาร(Pclass)

```python
sns.countplot(x='Pclass', hue = 'Survived', data=df_raw_all)
```

#### 3.4 สร้างกราฟ barplot นับจำนวนคน แบ่งตามผู้รอดชีวิตและไม่รอดชีวิตและแบ่งตามกลุ่มพี่น้องหรือสามีภรรยา(SibSp) โดยที่ข้อมูล SibSp ตั้งแต่ 1 คนขึ้นไป

```python
sns.countplot(x ='SibSp',hue = 'Survived', data=df_raw_all[df_raw_all['SibSp'] > 1])
```

#### 3.5 สร้างกราฟ barplot นับจำนวนคน แบ่งตามผู้รอดชีวิตและไม่รอดชีวิตและแบ่งตามกลุ่มพ่อแม่ลูก(Parch) โดยที่ข้อมูล Parch ตั้งแต่ 2 คนขึ้นไป

```python
sns.countplot(x ='Parch',hue = 'Survived', data=df_raw_all[df_raw_all['Parch'] > 2])
```

### 4. Factor Plot
#### 4.1  สร้างกราฟ Factor plot แบ่งตามผู้รอดชีวิตและไม่รอดชีวิต แกน X แสดง Pclass และแกน Y แสดง Age

```python
df_miss_all = df_raw_all.dropna(subset=['Age','Fare'], axis=0)
```

```python
g = sns.factorplot(x='Pclass', 
                   y='Age', 
                   data=df_miss_all, 
                   hue='Survived',  # Color by stage
                   col='Survived',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
```
